```python
import bs4
import urllib.request
```

```python
loc='https://www.zocdoc.com/search?address=New%20York%2C%20NY%2010027&insurance_carrier=1358&day_filter=AnyDay&gender=-1&language=-1&offset=0&insurance_plan=16150&reason_visit=75&after_5pm=false&before_10am=false&sees_children=false&sort_type=Default&dr_specialty=153&ip=160.39.9.117'
#create the page object using urllib
page=urllib.request.urlopen(loc)
#create the soup object
soup=bs4.BeautifulSoup(page,'html.parser')
```

```python
print(soup.prettify())
```

# ADDRESS

```python
#Getting the whole 'container' from the page
container=soup.find_all("div", class_="js-search-prof-list")[0]
```

```python
container.find_all()
```

```python
#Getting the addresses from that container
container.find_all("div", class_='js-search-prof-row-address')
```

```python
#Trying to get rid of second div
address = container.find("div", class_='js-search-prof-row-address')
print(address)
```

```python
for a in address.find_all("div"):
    a.decompose()
    
print(address.get_text())
```

```python
address2 = container.find_all("div", class_='js-search-prof-row-address')
print(address2)
```

```python
for address in address2:
    for a in address.find_all("div"):
        a.decompose()
    
for address in address2:
    print(address.getText())
```
